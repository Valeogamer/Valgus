```python
import zipfile
import os

# Путь к вашему архиву
zip_file_path = '/content/drive/MyDrive/NewValgus/2.KeyPoint/KeyPoint534/FootPronation.v2i.yolov5pytorch.zip'

# Папка, в которую нужно извлечь содержимое архива
extract_folder = '/content/Dataset/'

# Создание папки для извлечения, если ее нет
if not os.path.exists(extract_folder):
    os.makedirs(extract_folder)

# Извлечение содержимого архива
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extract_folder)

print("Архив успешно разархивирован.")
```
